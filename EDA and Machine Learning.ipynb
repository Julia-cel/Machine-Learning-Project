```python
import pandas as pd
import sklearn as sk
```

```python
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/clientes.csv')
display(df)
```

```python
df.info()
```

```python
#Quero prever a coluna score de crédito
#Toda IA só consegue trabalhar com números
#Coluna profissao,mix_credito,comportamento_pagamento são textos e tem que ser transformadas em números para
# a IA conseguir trabalhar com elas, a única coluna que se mantém como texto é a que queremos prever.
```

```python
#LabelEncoder para atribuir numeros a variaveis de texto
from sklearn.preprocessing import LabelEncoder

codificador_Prof= LabelEncoder()
df['profissao']= codificador_Prof.fit_transform(df['profissao'])

codificador_Credito= LabelEncoder()
df['mix_credito']= codificador_Credito.fit_transform(df['mix_credito'])

codificador_Pagamento= LabelEncoder()
df['comportamento_pagamento']= codificador_Pagamento.fit_transform(df['comportamento_pagamento'])
```

```python
# x= Colunas(dados) que vou usar pra fazer a previsão
# y= coluna(dados) que eu quero prever= score_credito > nunca usar a a coluna que vai prever para fazer a previsão
# Não pode prever duas coisas(só um Y)
y= df['score_credito']
x= df.drop(columns=['score_credito', "id_cliente"])
```

```python
# Função para separar parte do dataset para treino e a outra para teste. O treino serve para IA aprender a fazer previsão
# O teste é como uma prova para verificar se ela aprendeu

from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste= train_test_split(x, y) #Default order
#Parameter: test_size=0.3(30% teste, 70% treino)
#Passos para criar o modelo de IA:
#1: importar a IA, 2: Criar a IA, 3: Treinar a IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
#Criar a IA (por isso o parenteses fica em branco para preparar para receber treinamento):
modelo_ArvoreDecisão= RandomForestClassifier()
modelo_KNN= KNeighborsClassifier()
#Treinar a IA (fit):
modelo_ArvoreDecisão.fit(x_treino, y_treino)
modelo_KNN.fit(x_treino, y_treino)
#Depois de treinada, fazer previsões (predict):
previsoes_ArvoreDecisão= modelo_ArvoreDecisão.predict(x_teste)
previsoes_KNN= modelo_KNN.predict(x_teste)
from sklearn.metrics import accuracy_score
print( accuracy_score(y_teste, previsoes_ArvoreDecisão))
print( accuracy_score(y_teste, previsoes_KNN))
```
